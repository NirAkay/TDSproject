**Balancing Data** -<br>
We will try here to normalize the data to handle the problem of the extreme values that effects on our prediction.

```python
def norm_data_fumction(df):
  mean_scored_by = df['Scored By'].mean()
  std_scored_by = df['Scored By'].std()
  df['Scored By'] = (df['Scored By'] - mean_scored_by) / std_scored_by

df_piped = pd.DataFrame(anime_full_dataset)
norm_anime = df_piped.copy()

norm_data_fumction(norm_anime)
_, _, _, _, _ = run_prediction(norm_anime, 'Scored By')

anime_data_without_genres = norm_anime.drop(columns=["Genres"])

dtf_corr = anime_data_without_genres.corr(method="pearson").loc[["Scored By"]]
fig, ax = plt.subplots(figsize=(15, 2))
sns.heatmap(dtf_corr, annot=True, fmt='.2f', cmap="YlGnBu", cbar=True, linewidths=0.5, ax=ax)
plt.show()
```

We can see that our efforts improve the model because the **MAPE** reduced to 4.547

**Feature Engineering** -<br>
we will try here to create new columns so they will help us to improve our model and that because we want to utilize them and we are taking all the diffrent types of `Geners` and create from each one of them a new boolean feature. if in the original `Genres` feature anime has a list of genres that represent it for example: `Action, Drama, Comedy` so in the new features the anime get 1 in the features `Action`, `Drama`, `Comedy` and 0 in the others.

```python
def add_feature_fun(df):
  genres_split = df['Genres'].str.get_dummies(sep=', ')
  return pd.concat([df, genres_split], axis=1)

add_feature_anime = df_piped.copy()
add_feature_anime = add_feature_fun(add_feature_anime)
y_test_source, prediction_source, model_source, X_train_source, X_test_source = run_prediction(add_feature_anime, 'Scored By')

add_feature_anime = add_feature_anime.drop(columns=["Genres"])

dtf_corr = add_feature_anime.corr(method="pearson").loc[["Scored By"]]
fig, ax = plt.subplots(figsize=(15, 2))
sns.heatmap(dtf_corr, annot=True, fmt='.2f', cmap="YlGnBu", cbar=True, linewidths=0.5, ax=ax)
plt.show()
```

Unfortunatly we don't see any improvments and that's because our new features are not well corelated to `Scored By`. It looks like our linear model couldn't find a pattern from genres.

**Original Model**

```python
original_dataset = df_piped.copy()
y_test_source_original, prediction_source_original, model_source_original, X_train_source_original, X_test_source_original = run_prediction(original_dataset, 'Scored By')

original_dataset = original_dataset.drop(columns=["Genres"])

dtf_corr = original_dataset.corr(method="pearson").loc[["Scored By"]]
fig, ax = plt.subplots(figsize=(15, 2))
sns.heatmap(dtf_corr, annot=True, fmt='.2f', cmap="YlGnBu", cbar=True, linewidths=0.5, ax=ax)
plt.show()
```

**Improved Model**

```python
improved_model_anime = df_piped.copy()
improved_model_anime = improved_model_anime[(improved_model_anime['Scored By'] >= mean - 2 * std) & (improved_model_anime['Scored By'] <= mean + 2 * std)]
improved_model_anime = add_feature_fun(improved_model_anime)
norm_data_fumction(improved_model_anime)
y_test_source_improved, prediction_source_improved, model_source_improved, X_train_source_improved, X_test_source_improved = run_prediction(improved_model_anime, 'Scored By')

improved_model_anime = improved_model_anime.drop(columns=["Genres"])

dtf_corr = improved_model_anime.corr(method="pearson").loc[["Scored By"]]
fig, ax = plt.subplots(figsize=(15, 2))
sns.heatmap(dtf_corr, annot=True, fmt='.2f', cmap="YlGnBu", cbar=True, linewidths=0.5, ax=ax)
plt.show()
```

*It can be observed that **MAPE** dropped to 3.293 instead of 41.361.
***R squared** increased from 0.202 to 0.395.

Because we normalized the data, other parameters also changed, but we can't know how much from the change was because the model is better and how much is because of the normalization.

```python
from scipy.stats import ks_1samp, norm

basic_residuals = y_test_source_original - prediction_source_original
improved_residuals = y_test_source_improved - prediction_source_improved

fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=False)

axes[0].scatter(prediction_source_original, basic_residuals, alpha=0.6, label='Basic Model', color='purple')
axes[0].axhline(0, color='red', linestyle='--', label='Zero Error')
axes[0].set_title('Residual Plot - Basic Model')
axes[0].set_xlabel('Predicted Values')
axes[0].set_ylabel('Residuals (Actual - Predicted)')
axes[0].legend()
axes[0].grid(True)

axes[1].scatter(prediction_source_improved, improved_residuals, alpha=0.6, label='Improved Model', color='orange')
axes[1].axhline(0, color='red', linestyle='--', label='Zero Error')
axes[1].set_title('Residual Plot - Improved Model')
axes[1].set_xlabel('Predicted Values')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()

fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

sns.histplot(basic_residuals, kde=True, color='purple', label='Basic Model', alpha=0.6, ax=axes[0])
axes[0].set_title('Error Distribution - Basic Model')
axes[0].set_xlabel('Residuals (Actual - Predicted)')
axes[0].set_ylabel('Frequency')
axes[0].legend()
axes[0].grid(True)

sns.histplot(improved_residuals, kde=True, color='orange', label='Improved Model', alpha=0.6, ax=axes[1])
axes[1].set_title('Error Distribution - Improved Model')
axes[1].set_xlabel('Residuals (Actual - Predicted)')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

### Residual plots
We can see that in both plots the scatter is roughly the same, but in the improved graph the residuals are more tightly centered around the zero error line, which implies reduced variance and it has more consistent prediction range. Because we normalized our data, it sounds pretty obvious, but even when we try to account for that, the residuals in the improved model still look better. The std is equal to roughly 140,000 (see cell above). Now let's look at both graphs. When we look at the most negative prediction values, we can see that the residuals in the left graph, almost touch 250,000 which is almost 2 std while in the improved graph they almost don't cross the 1 std line. In addition, when we look at the higher prediction values, in both graphs, the error goes up, but in our improved model it doesn't go above 5 std, and in the basic model it reaches 10 stds.
### Error Distribution
In the error distribution we think that our model also improved, because there are less high errors, even when we compensate for the normalization (it can be seen in the graph that the basic model as a much longer tail). And the errors are also much more normalised. This can be seen visually and we will also present the ks value in the cell below

```python
basic_ks_stat, basic_ks_pvalue = ks_1samp(basic_residuals, cdf=norm.cdf, args=(np.mean(basic_residuals), np.std(basic_residuals)))
print(f"Basic Model KS Statistic: {basic_ks_stat}, P-value: {basic_ks_pvalue}")

improved_ks_stat, improved_ks_pvalue = ks_1samp(improved_residuals, cdf=norm.cdf, args=(np.mean(improved_residuals), np.std(improved_residuals)))
print(f"Improved Model KS Statistic: {improved_ks_stat}, P-value: {improved_ks_pvalue}")
```

Here we can see that the improved model is indeed much more normal. The P values increased but it's still so low that it doesn't matter.

## **Analysis**
Let's analyze what the improved model reveals about the data.

### Feature significance
To predict the scored by value, we learned that the most significant columns are Popularity, Score and Rank. Popularity being the most influential one after the improvements.
We though that the genre columns would have more impact, but our model didn't change significantlly after adding the columns.

### Biases and trends
We have a strong correlation between our numerical columns. For example as more people score an anime, it means it's more popular, thus having a lower ranking and popularity value. Also, more popular anime will probably have an higher score, and higher favorites and members value. That's why we had to remove the favorites and members columns, so our predictions won't be so good from the start.

### Data insights
We didn't expect that removing outliers from the lower values will improve our predictions as much

## **Our findings**
Our most insightful findings are:
- Anime with a Manga source, were scored by more people, thus being more popular
- Regular anime that is aired on TV, not specials or movies, tend to be a lot more popular, with a big difference.
